```python
# Função definidora do rotor_inicial
def define_posicao_do_rotor(numero):
    rotor_inicial = []
    if numero == 0:
        rotor_inicial = list(range(26))
    elif numero == 25:
        rotor_inicial = list(range(25, -1, -1))
    else:
        rotor_inicial = list(range(numero, 26))
    for i in range(0, numero):
        rotor_inicial.append(i)

    return rotor_inicial
```
