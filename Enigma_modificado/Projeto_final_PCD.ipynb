```python
# FunĂ§ĂŁo definidora do rotor_inicial
def define_posicao_do_rotor(numero):
    rotor_inicial = []
    if numero == 0:
        rotor_inicial = list(range(140))
    elif numero == 138:
        rotor_inicial = list(range(139, -1, -1))
    else:
        rotor_inicial = list(range(numero, 140))
    for i in range(0, numero):
        rotor_inicial.append(i)

    return rotor_inicial
```

```python
# FunĂ§ĂŁo para inverter o rotor
def inverso_rotor(rotor):
    inv = []
    for i in range(140):
        for j in range(140):
            if rotor[j] == i:
                inv.append(j)
    return inv
```

```python
texto=(input("Digite a palavra que vocĂȘ quer codificar"))
posicao_inicial_do_rotor=(int(input("Defina a posiĂ§ĂŁo inicial do rotor 1. Deve ser um nĂșmero de 0 a 25")))
posicao_inicial_do_rotor2=(int(input("Defina a posiĂ§ĂŁo inicial do rotor 2. Deve ser um nĂșmero de 0 a 25")))
posicao_inicial_do_rotor3=(int(input("Defina a posiĂ§ĂŁo inicial do rotor 3. Deve ser um nĂșmero de 0 a 25")))
conexĂŁo_entre_caracteres=(list(input("Quais caracteres vocĂȘ quer conectar?")))
posiĂ§ĂŁo_chave2=(int(input("Defina a posiĂ§ĂŁo chave do rotor 2. Deve ser um nĂșmero de 0 a 25")))
posiĂ§ĂŁo_chave3=(int(input("Defina a posiĂ§ĂŁo chave do rotor 3. Deve ser um nĂșmero de 0 a 25")))

#Definindo as equivalĂȘncias
dicionario_de_equivalĂȘncia={}
if conexĂŁo_entre_caracteres:
    for t in range(0,len(conexĂŁo_entre_caracteres),2):
       if t+1 < len(conexĂŁo_entre_caracteres):
            a=conexĂŁo_entre_caracteres[t]
            b=conexĂŁo_entre_caracteres[t+1]
            dicionario_de_equivalĂȘncia[a]=b
            dicionario_de_equivalĂȘncia[b]=a

#Definindo dicionĂĄrios
dicionario_caracter_para_numero={}
contador = 0
for i in range(33, 126):
    caractere = chr(i)
    dicionario_caracter_para_numero[caractere] = contador
    contador += 1

# Letras com acento e espaĂ§o
acentos = [
    'ĂĄ','Ă ','Ăą','ĂŁ','Ă€',
    'Ă©','Ăš','ĂȘ','Ă«',
    'Ă­','ĂŹ','Ăź','ĂŻ',
    'Ăł','ĂČ','ĂŽ','Ă”','Ă¶',
    'Ăș','Ăč','Ă»','ĂŒ',
    'Ă§',
    'Ă','Ă','Ă','Ă','Ă',
    'Ă','Ă','Ă','Ă',
    'Ă','Ă','Ă','Ă',
    'Ă','Ă','Ă','Ă','Ă',
    'Ă','Ă','Ă','Ă',
    'Ă',' '
]
for letra in acentos:
    if letra not in dicionario_caracter_para_numero:
        dicionario_caracter_para_numero[letra] = contador
        contador += 1

dicionario_numero_para_caracter={v:k for k,v in dicionario_caracter_para_numero.items()}

tam = len(dicionario_caracter_para_numero)

# definindo o refletor
refletor=list(range(tam))
for i in range(0, tam-1, 2):
    refletor[i]=i+1
    refletor[i+1]=i

#definindo rotores
rotor_1=define_posicao_do_rotor(posicao_inicial_do_rotor)
rotor_2=define_posicao_do_rotor(posicao_inicial_do_rotor2)
rotor_3=define_posicao_do_rotor(posicao_inicial_do_rotor3)

numero_de_voltas_1=0
numero_de_voltas_2=0
numero_de_voltas_3=0

rotor_1_inv=inverso_rotor(rotor_1)
rotor_2_inv=inverso_rotor(rotor_2)
rotor_3_inv=inverso_rotor(rotor_3)

#Aplicando as equivalĂȘncias
texto_com_equivalencia=[]
for v in texto:
    if v in dicionario_de_equivalĂȘncia:
        texto_com_equivalencia.append(dicionario_de_equivalĂȘncia[v])
    else:
        texto_com_equivalencia.append(v)

#passando mensagem para nĂșmeros
lista_palavra_em_numero=[]
for j in texto_com_equivalencia:
    lista_palavra_em_numero.append(dicionario_caracter_para_numero[j])

#Criptografia
lista_numerica_criptografada=[]
for z in lista_palavra_em_numero:
    numero_de_voltas_1 += 1


    if numero_de_voltas_1 % tam == posiĂ§ĂŁo_chave2:
        numero_de_voltas_2 += 1

    if numero_de_voltas_2 % tam == posiĂ§ĂŁo_chave3:
        numero_de_voltas_3 += 1

    #Atualiza rotores
    rotor_1=define_posicao_do_rotor(posicao_inicial_do_rotor + numero_de_voltas_1)
    rotor_2=define_posicao_do_rotor(posicao_inicial_do_rotor2 + numero_de_voltas_2)
    rotor_3=define_posicao_do_rotor(posicao_inicial_do_rotor3 + numero_de_voltas_3)

    rotor_1_inv=inverso_rotor(rotor_1)
    rotor_2_inv=inverso_rotor(rotor_2)
    rotor_3_inv=inverso_rotor(rotor_3)

    #Rotores na ida
    valor_1=rotor_1[z]
    valor_2=rotor_2[valor_1]
    valor_3=rotor_3[valor_2]

    valor_refletido=refletor[valor_3]

    #Rotores na volta
    valor_4=rotor_3_inv[valor_refletido]
    valor_5=rotor_2_inv[valor_4]
    valor_final=rotor_1_inv[valor_5]

    lista_numerica_criptografada.append(valor_final)

#Passando de nĂșmero para letra
lista_para_saida=[]
for t in lista_numerica_criptografada:
    resultado=dicionario_numero_para_caracter[t]
    if resultado in dicionario_de_equivalĂȘncia:
        resultado=dicionario_de_equivalĂȘncia[resultado]
    lista_para_saida.append(resultado)
mensagem_final="".join(lista_para_saida)
print(mensagem_final)
```
