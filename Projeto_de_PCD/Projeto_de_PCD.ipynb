#### Os Rotores ####



A Enigma possuía alguns mecanismos para realizar a criptografia, o primeiro deles eram chamados de rotores, cada rotor funcionava fazendo a relação de um par de letras onde uma representava a letra de entrada e a outra de saída, esse par dependia de qual configuração inicial do rotor que era escolhida, e devia ser previamente combinada entre quem manda e recebia a mensagem.

Como a Enigma possuia três rotores, a letra de saída do rotor anterior era a de entrada do próximo rotor, como no exemplo a seguir onde o usuário digita a letra "a":

A → G → F → T

Esse seria o caminho que a letra percorreria, se transforma para G no primeiro rotor, para F no segundo e para T no terceiro na configuração inicial combinada, ao alterar as condições inicias o caminho poderia ser outro como:

A → Y → B → K

Para representar esse mecanismo no projeto foi utilizada a estrutura de listas onde um cada número de 0 a 25 representa uma letra do alfabeto na entrada e sua posição na lista(Index) indicava sua letra de saída.

0 = a

1 = b

2 = c

3 = d

4 = e

e assim por diante....

O index de cada um destes números indica a letra de saída no mesmo padrão como por exemplo:

Na configuração onde o número 0 está no index 4:

Número 0, então letra de entrada = a

Index 4, então letra de saída = e

Então o caminho da letra "a" com um rotor na configuração descrita é:

A → E

No rotor seguinte a letra entra como 4 ou seja como letra "e" e o processo se repete.

Na Enigma a cada tecla que era digitada os rotores giravam e mudavam a relação entre entrada e saída, esse padrão de rotação era pré-definido ou seja, máquinas com mesma configuração inicial seguiam o mesmo caminho e a cada vez que uma letra era digitada ela percorria um caminho diferente nos rotores e poderia ser representada por uma letra diferente na criptografia


```python
def define_posicao_do_rotor_ev(numero):
    """Esta função define um rotor baseado em um número, este número
    é escolhido pelo usuário e será a letra que estará no index [0]"""
    rotor_posicionado = []
    if numero == 0:
        rotor_posicionado = list(range(26))
    elif numero == 25:
        rotor_posicionado = list(range(25, -1, -1))
    else:
        rotor_posicionado = list(range(numero, 26))
    for i in range(0, numero):
        rotor_posicionado.append(i)

    return rotor_posicionado
```

#### Os Refletores ####

A Enigma possuía também uma propriedade importante para seu funcionamento, ao criptografar uma mensagem e envia-lá a pessoa do outro lado precisa conhecer as configurações iniciais adotadas para conseguir digitar a mensagem e descriptografa-la, mas como uma mesma máquina criptografa e descripctografa?

Essa facilidade se deve graças aos refletores!!!

Eles eram os responsáveis por após as letras passarem pelos rotores, elas serem "refletidas" e fazerem o caminho contrário! Isso fazia com que o porecesso de cirptografia fosse simétrico e toda mensagem pudesse ser passada e recebida com a mesma configuração! Na Enigma ele atuava fechando o circuito elétrico e fazendo com que se "y" → "q", repetindo o mesmo processo "q" → "y".

Os refletores ajudavam muito na praticidade de enviar mensagens, mas deixavam um ponto fraco claro, uma letra não poderia se transformar nela mesma! A primeira vista parece apenas uma possibilidade a menos e que isso não irá afetar tanto assim a criptografia, mas esse ponto foi muito importante para a quebra da eficiência da comunicação Alemã!

```python
"""Função que é responsável por inverter o rotor, para ser usado em função
do processo de reflexão"""
def inverso_rotor_ev(rotor):
    inv = []
    for i in range(26):
        for j in range(26):
            if rotor[j] == i:
                inv.append(j)
    return inv
```

### "The plugboard" ###



Outro mecanismo utilizado pela enigma é o chamado "The plugboard", ele era constituído por um painel em que cada letra poderia ser ligada a outra por intermédio de um fio, na máquina isso trazia um efeito que ajudava a aumentar a segurança e o número de possibilidades as letras selecionadas se comportavam como as outras como por exemplo:

Se algúem ligasse as letras "b" e "g", as letras ligadas mudariam seu comportamento onde a letra "b" se comportaria com "g" e a letra "g" se comportaria como "b". 

Então na palavra gabriel se ao final dos rotores e refletores a palavara virasse "fbcqj fk" sem o The plugboard, com o The plugboard nas letras "b" e "g" a palavra seria "cgfqj fk"

Como são 26 letras, o números máximo de pares é 13 e o máximo de pares possíveis é de 325, porém ao analisar quantas configurações possíveis o painel inteiro pode representar esse número cresce muito! A Enigma era muito usada com 10 pares e 6 letras sobrando, dessa forma o numero de painéis possíveis é de aproximadamente 150 trilhões!!!

## A Enigma Verossímil ##

Nas céulas abaixo está a Enigma Verossímil! Para utiliza-la basta rodar as células e se divertir!
Porém primeiro é necessário entender como ela foi preparada para receber os imputs:

**1°**: Você deve digitar a mensagem que queira criptografar, a Enigma não aceitava espaços, mas para seu conforto pode espaçar o seu texto normalmente, porém cuidado! Não digite letras Maiúsculas, caracteres especiais ou espaços!

**2°, 3° e 4°**: Nestes imputs você deverá colocar uma posição entre 0 e 25 para seus rotores. No 2° imput para o primeiro rotor, no 3° input para o segundo rotor e no 4° input para o terceiro rotor.

**5°**: Aqui você deve digitair quais equivalências deseja estabelecer entre as letras, coloque-as minúsculas e sem espaço entre elas ou entre os pares, lembre-se que sempre deve se conectar uma letra a outra, ou seja o número de letras é sempre par! Uma letra não pode fazer duas conexões. Como as conexões são em ambos os sentidos o código já faz isso por você, então não é necessário digitar um par e em seguida o mesmo par de ordem invertida.

**6° e 7°**: Aqui você deve colocar o número de voltas relativas para cada rotor girar, o rotor 1 já está configurado para girar a cada caracter, o número que você digitar será para definir a cada quantas voltas do 1° o segundo girará e a cada quantas voltas do 2° o terceiro girará!




```python
# Função definidora de rotor
def define_posicao_do_rotor_ev(numero):
    rotor_posicionado = []
    if numero == 0:
        rotor_posicionado = list(range(26))
    elif numero == 25:
        rotor_posicionado = list(range(25, -1, -1))
    else:
        rotor_posicionado = list(range(numero, 26))
    for i in range(0, numero):
        rotor_posicionado.append(i)

    return rotor_posicionado
```

```python
# Função para inverter o rotor
def inverso_rotor_ev(rotor):
    inv = []
    for i in range(26):
        for j in range(26):
            if rotor[j] == i:
                inv.append(j)
    return inv
```

```python
texto=(input("Digite a palavra que você quer codificar")).replace(" ","")
posicao_inicial_do_rotor=(int(input("Defina a posição inicial do rotor 1. Deve ser um número de 0 a 25")))
posicao_inicial_do_rotor2=(int(input("Defina a posição inicial do rotor 2. Deve ser um número de 0 a 25")))
posicao_inicial_do_rotor3=(int(input("Defina a posição inicial do rotor 3. Deve ser um número de 0 a 25")))
conexão_entre_caracteres=(list(input("Quais letras você quer conectar?")))
posição_chave2=(int(input("Defina a posição chave do rotor 2. Deve ser um número de 0 a 25")))
posição_chave3=(int(input("Defina a posição chave do rotor 3. Deve ser um número de 0 a 25")))

#Definindo as equivalências
dicionario_de_equivalência={}
for t in range(0,len(conexão_entre_caracteres),2):
    if t+1 < len(conexão_entre_caracteres):
        a=conexão_entre_caracteres[t]
        b=conexão_entre_caracteres[t+1]
        dicionario_de_equivalência[a]=b
        dicionario_de_equivalência[b]=a

#Definindo dicionários
dicionario_caracter_para_numero={
 'a':0,'b':1,'c':2,'d':3,'e':4,'f':5,
 'g':6,'h':7,'i':8,'j':9,'k':10,'l':11,
 'm':12,'n':13,'o':14,'p':15,'q':16,
 'r':17,'s':18,'t':19,'u':20,'v':21,
 'w':22,'x':23,'y':24,'z':25,
}

dicionario_numero_para_caracter={v:k for k,v in dicionario_caracter_para_numero.items()}

# definindo o refletor
refletor=list(range(26))
for i in range(0,26,2):
    refletor[i]=i+1
    refletor[i+1]=i

#definindo rotores
rotor_1=define_posicao_do_rotor_ev(posicao_inicial_do_rotor)
rotor_2=define_posicao_do_rotor_ev(posicao_inicial_do_rotor2)
rotor_3=define_posicao_do_rotor_ev(posicao_inicial_do_rotor3)

numero_de_voltas_1=0
numero_de_voltas_2=0
numero_de_voltas_3=0

rotor_1_inv=inverso_rotor_ev(rotor_1)
rotor_2_inv=inverso_rotor_ev(rotor_2)
rotor_3_inv=inverso_rotor_ev(rotor_3)

#Aplicando as equivalências
texto_com_equivalencia=[]
for v in texto:
    if v in dicionario_de_equivalência:
        texto_com_equivalencia.append(dicionario_de_equivalência[v])
    else:
        texto_com_equivalencia.append(v)

#passando mensagem para números
lista_palavra_em_numero=[]
for j in texto_com_equivalencia:
    lista_palavra_em_numero.append(dicionario_caracter_para_numero[j])

#Criptografia
lista_numerica_criptografada=[]
for z in lista_palavra_em_numero:
    numero_de_voltas_1 += 1

    if numero_de_voltas_1 % 26 == posição_chave2:
        numero_de_voltas_2 += 1

    if numero_de_voltas_2 % 26 == posição_chave3:
        numero_de_voltas_3 += 1

    #Atualiza rotores
    rotor_1=define_posicao_do_rotor_ev(posicao_inicial_do_rotor + numero_de_voltas_1)
    rotor_2=define_posicao_do_rotor_ev(posicao_inicial_do_rotor2 + numero_de_voltas_2)
    rotor_3=define_posicao_do_rotor_ev(posicao_inicial_do_rotor3 + numero_de_voltas_3)

    rotor_1_inv=inverso_rotor_ev(rotor_1)
    rotor_2_inv=inverso_rotor_ev(rotor_2)
    rotor_3_inv=inverso_rotor_ev(rotor_3)

    #Rotores na ida
    valor_1=rotor_1[z]
    valor_2=rotor_2[valor_1]
    valor_3=rotor_3[valor_2]

    valor_refletido=refletor[valor_3]

    #Rotores na volta
    valor_4=rotor_3_inv[valor_refletido]
    valor_5=rotor_2_inv[valor_4]
    valor_final=rotor_1_inv[valor_5]

    lista_numerica_criptografada.append(valor_final)

#Passando de número para letra
lista_para_saida=[]
for t in lista_numerica_criptografada:
    resultado=dicionario_numero_para_caracter[t]
    if resultado in dicionario_de_equivalência:
        resultado=dicionario_de_equivalência[resultado]
    lista_para_saida.append(resultado)
#Adicionando espaçamento padrão
lista_para_saida_formatada=[]
contador=0
for caractere in lista_para_saida:
    lista_para_saida_formatada.append(caractere)
    contador+=1
    if contador==5:
        lista_para_saida_formatada.append(" ")
        contador=0

mensagem_final="".join(lista_para_saida_formatada)
print(mensagem_final)
```

```python
# Função definidora do rotor_inicial
def define_posicao_do_rotor(numero):
    rotor_inicial = []
    if numero == 0:
        rotor_inicial = list(range(140))
    elif numero == 138:
        rotor_inicial = list(range(139, -1, -1))
    else:
        rotor_inicial = list(range(numero, 140))
    for i in range(0, numero):
        rotor_inicial.append(i)

    return rotor_inicial
```

```python
# Função para inverter o rotor
def inverso_rotor(rotor):
    inv = []
    for i in range(140):
        for j in range(140):
            if rotor[j] == i:
                inv.append(j)
    return inv
```

```python
texto=(input("Digite a palavra que você quer codificar"))
posicao_inicial_do_rotor=(int(input("Defina a posição inicial do rotor 1. Deve ser um número de 0 a 140")))
posicao_inicial_do_rotor2=(int(input("Defina a posição inicial do rotor 2. Deve ser um número de 0 a 140")))
posicao_inicial_do_rotor3=(int(input("Defina a posição inicial do rotor 3. Deve ser um número de 0 a 140")))
conexão_entre_caracteres=(list(input("Quais caracteres você quer conectar?")))
posição_chave2=(int(input("Defina a posição chave do rotor 2.")))
posição_chave3=(int(input("Defina a posição chave do rotor 3.")))

#Definindo as equivalências
dicionario_de_equivalência={}
if conexão_entre_caracteres:
    for t in range(0,len(conexão_entre_caracteres),2):
       if t+1 < len(conexão_entre_caracteres):
            a=conexão_entre_caracteres[t]
            b=conexão_entre_caracteres[t+1]
            dicionario_de_equivalência[a]=b
            dicionario_de_equivalência[b]=a

#Definindo dicionários
dicionario_caracter_para_numero={}
contador = 0
for i in range(33, 126):
    caractere = chr(i)
    dicionario_caracter_para_numero[caractere] = contador
    contador += 1

# Letras com acento e espaço
acentos = [
    'á','à','â','ã','ä',
    'é','è','ê','ë',
    'í','ì','î','ï',
    'ó','ò','ô','õ','ö',
    'ú','ù','û','ü',
    'ç',
    'Á','À','Â','Ã','Ä',
    'É','È','Ê','Ë',
    'Í','Ì','Î','Ï',
    'Ó','Ò','Ô','Õ','Ö',
    'Ú','Ù','Û','Ü',
    'Ç',' '
]
for letra in acentos:
    if letra not in dicionario_caracter_para_numero:
        dicionario_caracter_para_numero[letra] = contador
        contador += 1

dicionario_numero_para_caracter={v:k for k,v in dicionario_caracter_para_numero.items()}

tam = len(dicionario_caracter_para_numero)

# definindo o refletor
refletor=list(range(tam))
for i in range(0, tam-1, 2):
    refletor[i]=i+1
    refletor[i+1]=i

#definindo rotores
rotor_1=define_posicao_do_rotor(posicao_inicial_do_rotor)
rotor_2=define_posicao_do_rotor(posicao_inicial_do_rotor2)
rotor_3=define_posicao_do_rotor(posicao_inicial_do_rotor3)

numero_de_voltas_1=0
numero_de_voltas_2=0
numero_de_voltas_3=0

rotor_1_inv=inverso_rotor(rotor_1)
rotor_2_inv=inverso_rotor(rotor_2)
rotor_3_inv=inverso_rotor(rotor_3)

#Aplicando as equivalências
texto_com_equivalencia=[]
for v in texto:
    if v in dicionario_de_equivalência:
        texto_com_equivalencia.append(dicionario_de_equivalência[v])
    else:
        texto_com_equivalencia.append(v)

#passando mensagem para números
lista_palavra_em_numero=[]
for j in texto_com_equivalencia:
    lista_palavra_em_numero.append(dicionario_caracter_para_numero[j])

#Criptografia
lista_numerica_criptografada=[]
for z in lista_palavra_em_numero:
    numero_de_voltas_1 += 1


    if numero_de_voltas_1 % tam == posição_chave2:
        numero_de_voltas_2 += 1

    if numero_de_voltas_2 % tam == posição_chave3:
        numero_de_voltas_3 += 1

    #Atualiza rotores
    rotor_1=define_posicao_do_rotor(posicao_inicial_do_rotor + numero_de_voltas_1)
    rotor_2=define_posicao_do_rotor(posicao_inicial_do_rotor2 + numero_de_voltas_2)
    rotor_3=define_posicao_do_rotor(posicao_inicial_do_rotor3 + numero_de_voltas_3)

    rotor_1_inv=inverso_rotor(rotor_1)
    rotor_2_inv=inverso_rotor(rotor_2)
    rotor_3_inv=inverso_rotor(rotor_3)

    #Rotores na ida
    valor_1=rotor_1[z]
    valor_2=rotor_2[valor_1]
    valor_3=rotor_3[valor_2]

    valor_refletido=refletor[valor_3]

    #Rotores na volta
    valor_4=rotor_3_inv[valor_refletido]
    valor_5=rotor_2_inv[valor_4]
    valor_final=rotor_1_inv[valor_5]

    lista_numerica_criptografada.append(valor_final)

#Passando de número para letra
lista_para_saida=[]
for t in lista_numerica_criptografada:
    resultado=dicionario_numero_para_caracter[t]
    if resultado in dicionario_de_equivalência:
        resultado=dicionario_de_equivalência[resultado]
    lista_para_saida.append(resultado)
mensagem_final="".join(lista_para_saida)
print(mensagem_final)
```
